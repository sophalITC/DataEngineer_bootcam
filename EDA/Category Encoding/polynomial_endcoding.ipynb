# 11. Polynomial Encoding

Polynomial encoding, also known as polynomial regression or polynomial features, is a technique used in machine learning and statistics to capture nonlinear relationships between variables. It involves transforming the original features or predictors by creating new features that are derived from the powers of the original features.


```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
import category_encoders as ce
```

```python
df = pd.read_csv(
        'https://raw.githubusercontent.com/scikit-learn-contrib/categorical-encoding/master/examples/source_data/cars/car.data.txt'
    )
df
```

```python
df["class"].unique()
```

```python
X = df.reindex(columns=[x for x in df.columns.values if x != 'class'])
y = df.reindex(columns=['class'])
y = preprocessing.LabelEncoder().fit_transform(y.values.reshape(-1, ))
```

```python
y
```

```python
X.head()
```

```python
X_trans = ce.PolynomialEncoder().fit_transform(X,y)
X_trans.head()
```
